```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Import GDP per capita file
df_gdp = pd.read_csv('GDP_per_capita.csv')
df_gdp
```

```python
df_gdp.columns
```

```python
#drop unneeded columns
df_gdp.drop(columns=['Country Code', 'Indicator Name', 'Indicator Code',
       '1960', '1961', '1962', '1963', '1964', '1965', '1966', '1967', '1968',
       '1969', '1970', '1971', '1972', '1973', '1974', '1975', '1976', '1977',
       '1978', '1979', '1980', '1981', '1982', '1983', '1984', '1985', '1986',
       '1987', '1988', '1989', '1990', '1991', '1992', '1993', '1994', '1995',
       '1996', '1997', '1998', '1999'], axis=1, inplace = True)
```

```python
#check whether columns are dropped
df_gdp.columns
```

```python
#rename column for ease of use
df_gdp.rename(columns = {'Country Name':'country'}, inplace = True)
```

```python
# "Melt" the years into one column
df_gdp = df_gdp.melt(id_vars=["country"], var_name="year", value_name="gdp_per_capita")
df_gdp
```

```python
df_gdp.info()
```

```python
#set year to integer
df_gdp["year"] = df_gdp["year"].astype(int)
```

```python
df_gdp.set_index(["country", "year"], inplace = True)
df_gdp.sort_index(inplace=True)
```

```python
#Import Debt file
df_debt = pd.read_excel('Central_Government_Debt_Percentage_of_GDP.xlsx')
df_debt
```

```python
df_debt.columns
```

```python
#drop unneeded columns
df_debt.drop(columns=[1950,1951,1952,1953,1954,1955,1956,1957,1958,1959,
       1960, 1961, 1962, 1963, 1964, 1965, 1966, 1967, 1968,
       1969, 1970, 1971, 1972, 1973, 1974, 1975, 1976, 1977,
       1978, 1979, 1980, 1981, 1982, 1983, 1984,1985, 1986,
       1987, 1988, 1989, 1990, 1991, 1992, 1993, 1994, 1995,
       1996, 1997, 1998, 1999], axis=1, inplace=True)
```

```python
#check whether columns are dropped
df_debt.columns
```

```python
#rename column for ease of use
df_debt.rename(columns={'Central Government Debt (Percent of GDP)':'country'},inplace = True)
df_debt
```

```python
#drop last 2 rows
df_debt = df_debt.iloc[:-2]
df_debt
```

```python
# "Melt" the years into one column
df_debt = df_debt.melt(id_vars=["country"], var_name="year", value_name="percentage_of_gdp")
df_debt
```

```python
df_debt.set_index(["country", "year"], inplace = True)
df_debt.sort_index(inplace=True)
df_debt
```

```python
"United Kingdom" in df_debt.index.get_level_values("country")
```

```python
"Japan" in df_debt.index.get_level_values("country")
```

```python
"France" in df_debt.index.get_level_values("country")
```

```python
fig, ax = plt.subplots(figsize=(10,6))

# First axis (GDP per capita)
ax.plot(df_gdp.loc['United Kingdom'].index, df_gdp.loc['United Kingdom']["gdp_per_capita"], 
        label='GDP per capita', color='green')
ax.set_ylabel('GDP per capita in US Dollars')
ax.set_xlabel('Year')
ax.set_xlim(1998, 2027)
ax.set_ylim(25000, 55000)

# Second axis (Debt as % of GDP)
ax2 = ax.twinx()
ax2.plot(df_debt.loc['United Kingdom'].index, df_debt.loc['United Kingdom']["percentage_of_gdp"], 
         label='Debt as % of GDP', color='red', linestyle='dashed')
ax2.set_ylabel('Debt as a Percentage of GDP')
#ax2.set_xlabel('Year')
ax2.set_ylim(0, 120)

# Vertical spans
ax.axvspan(2005, 2006, color='red', alpha=0.5, label='Olympics Announcement')
ax.text(2005.5, 32500, 'Olympics Announcement', rotation=90, horizontalalignment='center', color='black')
ax.axvspan(2012, 2013, color='blue', alpha=0.5, label='London Olympics')
ax.text(2012.5, 35000, 'London Olympics', rotation=90, horizontalalignment='center', color='black')

# Combine legends from both axes
lines1, labels1 = ax.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax.legend(lines1 + lines2, labels1 + labels2, loc='upper left', fontsize = 9)

ax2.set_title('UK Economic Indicator Before and After Olympics', fontsize = 15)
plt.tight_layout()
plt.show()
```
